```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('MER_T12_06 2.csv')
```

```python
data['date'] = pd.to_datetime(data.YYYYMM.apply(str), format='%Y%m',errors='coerce')
```

```python
import seaborn as sns
sns.set()
```

```python
data.Value = data.Value.convert_objects(convert_numeric=True)
```

```python
data.Value.dtype
```

```python
data.dropna().Value.hist(bins=100)
```

```python
data.Description.unique()
```

```python
sources = data.Description.unique()[0:-1]
```

```python
plt.figure(figsize=(25,20))

for ind,source in enumerate(sources):
    plt.subplot(4, 2, ind+1)
    plt.hist(data.dropna()[data.dropna()['Description']==source]['Value'], bins=15, alpha=0.5, label=source)
    plt.xlabel('Million Metric Tons of Carbon Dioxide',fontsize=20)
    plt.ylabel('Counts',fontsize=20)
    plt.legend(loc='upper right',fontsize=10)
```

```python
fig = plt.figure(figsize=(25,20))
ax1 = fig.add_subplot(111)
plt.yscale('log')
plt.ylim([10**-3,10**3])

for ind,source in enumerate(sources):
    data.dropna()[data.dropna()['Description']==source].plot(x='date',y='Value',label=source, ax=ax1)
    plt.xlabel('Year',fontsize=25)
    plt.ylabel('Million Metric Tons of Carbon Dioxide',fontsize=25)

data.dropna()[data.dropna()['Description']=='Total Energy Electric Power Sector CO2 Emissions'].plot(x='date',y='Value',
                                                                                                     label='Total Energy Electric Power Sector CO2 Emissions', ax=ax1)
plt.xlabel('Year',fontsize=25)
plt.ylabel('Million Metric Tons of Carbon Dioxide',fontsize=25)
    
plt.legend(fontsize=18)
```

```python
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(111)
plt.yscale('log')
plt.ylim([10**-1,0.1*10**4])

y = []

total_co2_index = data.dropna()[data.dropna()['Description']=='Total Energy Electric Power Sector CO2 Emissions']['date'].tolist()

for ind,source in enumerate(sources[::-1]):
    selected = data.dropna()[data.dropna()['Description']==source]
    selected = selected.set_index('date')
    y.append(selected['Value'].loc[total_co2_index].tolist())

plt.stackplot(total_co2_index,y, labels=sources[::-1])  
plt.legend()
plt.xlabel('Year',fontsize=25)
plt.ylabel('Million Metric Tons of Carbon Dioxide',fontsize=25)
```

```python
all_gdp = pd.read_csv('usa_gdp_ppp.csv')
```

```python
usa_gdp = all_gdp[all_gdp['Country Name']=='United States']
```

```python
usa_gdp = usa_gdp.T.loc[list([str(i) for i in range(1990,2017)]),:]
usa_gdp.columns = ['GDP_PPP']
```

```python
usa_gdp.plot()
plt.ylabel('GDP PPP (constant 2011 international $)')
plt.xlabel('Year')
```

```python
usa_total_co2_electric = data.dropna()[data.dropna()['Description']=='Total Energy Electric Power Sector CO2 Emissions']
```

```python
usa_total_co2_electric['year'] = usa_total_co2_electric.date.apply(lambda x: x.year)
usa_total_co2_electric.head()
```

```python
co2_electric = usa_total_co2_electric.groupby('year')['Value'].mean().loc[range(1990,2017)]
```

```python
co2_electric = pd.DataFrame(co2_electric)
```

```python
co2_electric.index
```

```python
usa_gdp.index = [int(i) for i in usa_gdp.index]
```

```python
usa_gdp.index
```

```python
usa_gdp.head()
```

```python
combined = co2_electric['Value']/usa_gdp['GDP_PPP']
```

```python
fig, ax1 = plt.subplots(figsize=(10, 10))
combined.plot(kind='line',ax=ax1,color='r',lw=5)
plt.ylabel('Million Metric Tonnes of CO2 / GDP PPP (constant 2011 international $)',fontsize=15,color='r')
plt.yticks(color='r')
plt.xlabel('Year',fontsize=15)

ax2 = ax1.twinx()
usa_gdp.plot(ax=ax2,lw=5)
plt.ylabel('GDP PPP (constant 2011 international $)',fontsize=15,color='b')
plt.yticks(color='b')

ax2.get_legend().remove()

plt.xlabel('Year',fontsize=15)
```
